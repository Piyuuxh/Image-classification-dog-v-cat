```python
import tensorflow as tf
import numpy as np
```

```python
filename = input("Paste an image URL : ")
```

```python
#from IPython.display import Image
#Image(filename = 'dog.jpg', width = 224, height = 224)
```

```python
mobile = tf.keras.applications.mobilenet_v2.MobileNetV2()
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
img = image.load_img(filename, target_size = ( 224, 224))
```

```python
from typing_extensions import final
resized_image = image.img_to_array(img)
final_image = np.expand_dims(resized_image, axis = 0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
final_image.shape
```

```python
predictions = mobile.predict(final_image)
from tensorflow.keras.applications import imagenet_utils
results = imagenet_utils.decode_predictions(predictions)
```

```python
for res in results:
  for i in res:

    if 'cat' in i[1]:
      print("This is cat!")
      break

    else:
      print("This is dog!")
      break

plt.imshow(img)
```
